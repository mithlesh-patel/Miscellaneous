```python
import pandas as pd
location='/Users/mithleshpatel/Documents/Command Manager/'
```

```python
with open(location+'CmdMgrSuccessMetricsProperties.log') as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
content = [x.strip() for x in content]
content
```

```python
#f = open(location+'Delete_Unused_Metrics_Prod.scp', 'w')
#f2 = open(location+'Unused_Metrics1.txt', 'w')
  # python will convert \n to os.linesep

df = pd.DataFrame(columns=[['Metric','Location','Formula','Expression']])
for line in content:
    if line.count('|')==3:
        lines= line.split("GMT",1)[1] 
        substack = lines.split('|')
        #print substack
        df = df.append({'Metric':substack[0].strip(),'Location':substack[1].strip(),
                        'Formula':substack[2].strip(),
                       'Expression':substack[3].strip()},ignore_index=True)
        #f2.write(lines + '\n')
        #newline = 'DELETE METRIC ' + '"'+substack[0].strip() + '"' + ' IN FOLDER ' + '"' + substack[1].strip()\
        #+ '"' + ' FROM PROJECT "Badoo BI";'
        #f.write(newline +'\n')
        #print '****'
#f.close()
#f2.close()
df.to_csv(location+'MetricProperty.csv')
df.head()
```
