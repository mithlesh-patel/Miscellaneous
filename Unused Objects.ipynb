```python
import pandas as pd
location='/Users/mithleshpatel/Documents/Command Manager/'
```

```python
with open(location+'CmdMgrSuccessFlagTables.log') as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
content = [x.strip() for x in content]
content
```

```python
#f = open(location+'Delete_Unused_Metrics_Prod.scp', 'w')
#f2 = open(location+'Unused_Metrics1.txt', 'w')
  # python will convert \n to os.linesep

df = pd.DataFrame(columns=[['Table','Dependent_Object','Dependent_Attribute','Dep_Obj_Loc','Dep_Obj_Type']])
for line in content:
    if line.count('|')==4:
        lines= line.split("GMT",1)[1] 
        substack = lines.split('|')
        #print substack
        df = df.append({'Table':substack[0].strip(),'Dependent_Object':substack[2].strip(),
                        'Dependent_Attribute':substack[1].strip(),
                       'Dep_Obj_Loc':substack[3].strip(),'Dep_Obj_Type':substack[4].strip()},ignore_index=True)
        #f2.write(lines + '\n')
        #newline = 'DELETE METRIC ' + '"'+substack[0].strip() + '"' + ' IN FOLDER ' + '"' + substack[1].strip()\
        #+ '"' + ' FROM PROJECT "Badoo BI";'
        #f.write(newline +'\n')
        #print '****'
#f.close()
#f2.close()
df.to_csv(location+'Flag_Tables_By_Report.csv')
df
```

```python
df3=df2.pivot_table(index=['Table','Dependent_Attribute'],columns=['Dependent_Object'],aggfunc=len,fill_value=0)
```

```python
df3.to_csv(location+'tablesflag.csv')
```

```python
## add a row total to dft1 with the following steps

sum_row = df3.sum(axis=0) #get sum_row first
#print sum_row
dft1_sum=pd.DataFrame(data=sum_row).T #change it to a dataframe
#dft1_sum
dft1_sum=dft1_sum.reindex(columns=df3.columns) #line up the col index to dft1
#dft1_sum
dft1_sum.index = ['row_total'] #change row index to row_total
#dft1_sum
df4=df3.append(dft1_sum) # append the row to dft1
df4.to_csv(location+'FlagTablesCount.csv')
```
