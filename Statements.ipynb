```python
import xlrd
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
location='/Users/mithleshpatel/anaconda2/zipzap/Data_File/ICICI/'
dd=pd.DataFrame()
for f in glob.glob(location+"519109640_*.xls"):
    workbook = xlrd.open_workbook(f)
    for sheet in workbook.sheets():
        start_dates=[]
        cols=[] 
        for row in range(sheet.nrows):
            for column in range(sheet.ncols):
                if sheet.cell(row,column).value in ['DATE','MODE','PARTICULARS','DEPOSITS','WITHDRAWALS','BALANCE']:
                    start_dates.append(row)
                    cols.append(column)
    unq_rows=set(start_dates)
    unq_cols=set(cols)
    d_list=[]
    d_list.append([])
    
    for r in range(min(unq_rows)+1,sheet.nrows-1):
        cell_value = sheet.cell(r,1).value

        #print type(cell_value)
        
        if (isinstance(cell_value, unicode) and cell_value.count('-') == 2 and len(cell_value)==10):
            e_row=[]
            for c in unq_cols:
                e_row.append(sheet.cell(r,c).value)
            #print e_row
            d_list.append(e_row)
                
    
    d_list = d_list[1:]
    dd=dd.append(d_list,ignore_index=True)

                    
            
```

```python
dd.columns=['DATE','MODE','PARTICULARS','DEPOSITS','WITHDRAWALS','BALANCE']
dd.to_csv(location+'Statements.csv')
dd[dd['MODE'].str.contains("CHEQUE")]
```
